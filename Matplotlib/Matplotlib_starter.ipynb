```python
import pandas as pd
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', linestyle = 'dashed')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dotted')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', linestyle = 'dotted')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = '3')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', linestyle = 'solid', linewidth = '2')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = '3', marker = 'D', markersize = '10')
plt.plot(batsman['index'],batsman['RG Sharma'], color='red', linestyle = 'solid', linewidth = '2', marker = '+')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = '3', label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', linestyle = 'solid', linewidth = '2', label = 'Rohit')

plt.title("Rohit vs Virst Comparison")
plt.xlabel("Season")
plt.ylabel("Season")

plt.legend()
```
